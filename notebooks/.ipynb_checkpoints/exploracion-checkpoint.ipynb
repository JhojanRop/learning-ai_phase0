```python
import pandas as pd
```

```python
df = pd.read_csv("../data/dataset.csv")
print(df.head(5))
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
print(df[df["artists"].isnull()])
print(df[df["album_name"].isnull()])
print(df[df["track_name"].isnull()])
```

```python
df = df.dropna()
print(df.isnull().sum())
```

```python
print(df.dtypes)
```

```python
print(f"Popularity: {df['popularity'].min()} - {df['popularity'].max()}")
print(f"Duration: {df['duration_ms'].min()} - {df['duration_ms'].max()}")
print(f"Tempo: {df['tempo'].min()} - {df['tempo'].max()}")
print(f"Danceability: {df['danceability'].min()} - {df['danceability'].max()}")
print(f"Energy: {df['energy'].min()} - {df['energy'].max()}")
```

```python
stats = (
    df[[
        "popularity",
        "duration_ms",
        "danceability",
        "energy",
        "loudness",
        "mode",
        "speechiness",
        "acousticness",
        "instrumentalness",
        "liveness",
        "valence",
        "tempo",
        "time_signature"
    ]]
    .agg(["min", "max"])
    .T
)

print(stats)
```
