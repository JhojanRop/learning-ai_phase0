Se aplica minúscula al nombre de la canción y artista para eliminar repetidos como en el caso anterior.

```python
lower_df = df.copy()
lower_df['track_name'] = lower_df['track_name'].str.lower()
lower_df['artists'] = lower_df['artists'].str.lower()
# group2 = lower_df.groupby(["track_name", "artists"], as_index=False).agg(list)
```

```python
# group2.shape
```

```python
# group2.head(5)
```

```python
df_clean = lower_df.copy()
df_clean = df_clean.dropna()
df_clean = df_clean.drop(columns=["Unnamed: 0"])
```

```python
df_clean.groupby(["track_name", "artists"], as_index=False).agg(list)
```

```python
df_clean.dtypes
```

```python
df_clean = (
    df_clean.groupby(["track_name", "artists"], as_index=False).agg({
        'album_name': 'first',
        'popularity': 'max',
        'duration_ms': 'mean',
        'explicit': lambda x: x.mode().iloc[0],
        'danceability': 'mean'
    })
)
```
